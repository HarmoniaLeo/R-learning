```python
from sklearn import datasets
import pandas as pd

iris_datas = datasets.load_iris()
iris_datas = pd.DataFrame(iris_datas['data'])
```

```python
iris_datas
```

```python
iris_datas[0]=iris_datas[0].astype(int)
iris_datas
```

```python
def func(x):
    q25=x.quantile(0.25)
    q75=x.quantile(0.75)
    IQR=q75-q25
    return ((x>q75+1.5*IQR)|(x<q25-1.5*IQR)).any(axis=1).sum()

iris_datas.groupby(0).apply(func)
```

```python
def func(x):
    target_col = 1
    corrs = x.corr().abs()[target_col]
    return pd.Series([corrs.nlargest(2).index[1],corrs.nlargest(2).iloc[1]],index=["max_corr_col","max_corr"])

iris_datas.groupby(0).apply(func)
```

```python
target_cols = iris_datas.columns[1:]

def func(x):
    corrs = x[target_cols].corr().abs()
    return pd.Series(pd.Series(target_cols,index=target_cols).apply(lambda x:corrs[x].nlargest(2).index[1]))

iris_datas.groupby(0).apply(func)
```

```python
import math

def conb(n,m):
    return math.factorial(n)/(math.factorial(n-m)*math.factorial(m))


conb(20,10)/conb(100,10)
```
